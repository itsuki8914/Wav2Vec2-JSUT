```python
import pandas as pd
import pykakasi as pk
import yaml
import os, glob
import re
```

```python
tgt = 'hira'
val_size = 500
```

```python
tsv_file = glob.glob('*corpus*ja/*/ja/val*.tsv')[0]
tsv_file
```

```python
import pandas as pd
df = pd.read_csv(tsv_file, index_col=0, sep='\t')
df.head()
```

```python
mp3_folder = glob.glob('*corpus*ja/*/ja/*clip*')[0]
mp3_folder
```

```python
def add_path(path):
    path = path.replace('mp3', 'wav')
    new_path = './data/processed_clips/' + path
    #new_path = './'+ mp3_folder + '/' + path
    return new_path
```

```python
df['path'] = df['path'].map(add_path)
df.head()
```

```python
kks = pk.kakasi()
```

```python
def ret_tgt(batch):
    res = kks.convert(batch)
    ret = [res[i][tgt] for i in range(len(res))]
    return ''.join(ret)
```

```python
%%time
df[tgt] = df["sentence"].map(ret_tgt)
df.head()
```

```python
def remove_special_characters(batch):
    data = batch
    if type(data) == str:
        data = batch.translate(str.maketrans({chr(0xFF01 + i): chr(0x21 + i) for i in range(94)}))
        data = re.sub(r'\n', '', data)
        data = ''.join(char for char in data if char.isalnum())
    return data
```

```python
df[tgt] = df[tgt].map(remove_special_characters)
```

```python
vocab_df = []
for i in range(len(df)):
    for j in range(len(df[tgt].iloc[i])):
        vocab_df.append(df[tgt].iloc[i][j:j+1].lower())

vocab_list = sorted(list(set(vocab_df)))
vocab_dict = {v: k for k, v in enumerate(vocab_list)}
vocab_dict
```

```python
vocab_dict[" "] = len(vocab_dict)
vocab_dict["|"] = vocab_dict[" "]
del vocab_dict[" "]
vocab_dict["[UNK]"] = len(vocab_dict)
vocab_dict["[PAD]"] = len(vocab_dict)
len(vocab_dict)
```

```python
import json
with open(f'vocab_{tgt}.json', 'w') as vocab_file:
    json.dump(vocab_dict, vocab_file)
```

```python
val = df[:val_size]
train = df[val_size:]
len(train)
```

```python
train.to_csv(f'train_{tgt}_wav.csv')
val.to_csv(f'val_{tgt}_wav.csv')
```
